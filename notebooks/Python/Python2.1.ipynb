## If-Else Statements

If-Else statements are one of the first steps to allowing our programs to make decisions.

```python
varx = True

if varx == True:
    print('Variable X is True')
else:
    print('Variable X is not True')
```

It is *very* important to make note of the spacing here.

Both *If-Else* statements and *Functions* use indentation has a way to identify the various parts.

Typically code will be indented either two or four spaces.

> [PEP 8](https://pep8.org/) the Python Style Guide

> The *PEP 8* guidelines are an excellent resource to read through once, simply for awareness!

Also notice how we *ask* our *if* question with the double equals sign.

In Python, a single equals sign set a variable.

A double equals sign affirms that two objects have the *same value*; which is not the same as being the same.

[More Notes](https://stackoverflow.com/questions/15008380/double-equals-vs-is-in-python)

> In JavaScript there is a triple equals sign, to check if two variables have the same value and are the same type! Which again, is not the same as being the same.

There is a larger to be discussed here, though perhaps it concerns JavaScript a bit more. And that is whether a variable is an object, or a representation of an object.

Whenevery anything is written by a computer, binary code is changed at byte-level.

When you define a variable `varx = 3`, that piece of information is written into bytes and stored at a specific location.

Two things can happen from here: 

1) this piece of code can be changed. `varx = 5`

2) the code can be referenced, and the reference changed. `varz = varx + 5`

I am simply adding this reference here as an extra nugget of info for context.

These concepts are more at play when using JavaScript than in Python.

## If-ElseIf-Else

The binary option of only *if* or *else* is simply not tenable for application in isolation; we need to be able to have more options.

Enter the *ElseIf* or *elif* option!

```python
varx = 'green'

if varx == 'yellow':
    print('Variable X is Yellow')
elif varx == 'blue':
    print('Variable X is Blue')
else:
    print('Variable X is neither Yellow nor Blue')
```

We have changed Variable X to use a string instead of Boolean statement, and the *else* statement serves are our 'catch-all' reply.

> It is a best practice to make sure we always conclude our statements and functions with some indication of the resolution taken by the code.
