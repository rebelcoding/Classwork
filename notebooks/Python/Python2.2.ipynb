## Lists

List are collections of items bound by square brackets, we know this.

```python
newList = ['red', 'yellow', 'blue', 'black', 'white']
```

And through our exploration earlier with indexing strings, we can understand how to index lists!

```python
newList = ['red', 'yellow', 'blue', 'black', 'white']
print(newList[2]) # what color does this command print?
print(newList[2:]) # what colors does this command print?
print(newList[:2]) # what colors does this command print?
```

Mind you, lists don't just have to contain strings.

List can contain many different objects:

```python
newBool= True
newString = 'Gigawatt'
newList = ['alphnums', 1234, False, 1.234]
newColors = ['red', 'yellow', 'blue', 'black', 'white']
baseList = [newBool, newString, newList, newColors]
print(baseList) # what does this command print?

# You can add lists:

compoundList = newList + newColors
print(compoundList)

# We use the .append() function to add item to a list

newColors.append('green')
print(newColors)

# Did compoundList change though?
print(compoundList)

# We use the .pop() function to remove items from a list

poppedItem = newColors.pop(5)
print(poppedItem)
print(newColors)
```



## For Loops

For-loops are one way to process lists.

```python
newList = ['red', 'yellow', 'blue', 'black', 'white']

for item in newList:
    print(item)
```
