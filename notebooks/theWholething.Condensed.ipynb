## HTML & CSS

HTML is the basic framing for webpages.

Even though *most* websites use technologies beyond HTML, the information is often presented through HTML, or a similarly structured frame :)

Below is an example containing, practically, the least code needed to display a page using HTML.


```html
<html>
    <head>
        <title>Home</title>
    </head>
    <body>
        <h1>Welcome to SF</h1>
    </body>
</html>
```

We have what we will call *tags*, which are *nested* within each other. 

The words that I have italicized are worth remembering as concepts.

We have an `<html>` tag that contains two other tags, `<head>` and `<body>`. 

Each of these contain yet another tag.

And finally, there is some text.

You can view this page in action [here](//screamfreely.github.io).

So besides everything, what else are we missing??

CSS for one ...

**Cascading Style Scripts** are *scripts that style our code*; we'll explain the cascading part in a moment.

If I want to change the color of an item, I would use CSS.

There are many different ways to use CSS, and almost like JavaScript, the number of CSS libraries seems to always be growing.

Regardless, the basics are as follows:

#### Inline

Here we add the styling code directly within the tag to which it is to be applied.

We add a *style* property to the tag, and then include our preferred augmentation.

```html
<html>
    <head>
        <title>Home</title>
    </head>
    <body>
        <h1 style="color: #42FD23; font-size: 5em;">Welcome to SF</h1>
    </body>
</html>
```
#### Embedded

What makes CSS nifty, is that you can use it to make adjustments to whole groups or types of tags!

Instead of including a *style* property on the `<h1>` tag, we use the `<style>` tag ~ in this tag we are able to list the tags we want to affect, and define our preferences.

```html
<html>
    <head>
        <title>Home</title>
        <style>
            h1 {
                color: #42FD23;
            }
        </style>
    </head>
    <body>
        <h1>Welcome to SF</h1>
    </body>
</html>
```
#### Embedded + Class

Just like before, but now we are going to take a look at classes.

What if you don't want to affect every `<h1>` tag?

We use classes! Classes are heavily used in CSS and styling libraries.

The *class* property does allow for more than parameter, separated by a space.

```html
<html>
    <head>
        <title>Home</title>
        <style>
            h1 {
                color: #42FD23;
            }
            .headline {
                text-decoration: underline;
            }
        </style>
    </head>
    <body>
        <h1 class="headline other-class">Welcome to SF</h1>
    </body>
</html>
```

#### Imported

Lastly we imagine that for an entire application there is going to be a whole heap of CSS code. And having it all in the same file as the base file will be ridiculous!

Entonces, we are going to write a completely separate file, and *import* it into our base file.

We do this using a `<link>` tag, that contains two properties, *rel* and *href*. *rel* defines the relationship of the file being consumed; and *href* provides the address at which the file can be found, *hyperlink-reference*.

```html
<html>
    <head>
        <title>Home</title>
        <link rel="stylesheet" href="style.css"/>
    </head>
    <body>
        <h1>Welcome to SF</h1>
    </body>
</html>
```

**Cascading** refers to how instructions are interpreted.

What if the *style.css* scripts says `<h1>` tags should all be green, but a style property on the page specifies that this specific `<h1>` tag should be blue??

*Most of the time* the page, or component, specific parameter/property/attribute will take precedence in styling the tag. And this is attributed to the cascading nature of style scripts :)

##### 3) Nested properties
Ok, I am cheating here and combining to ideas into one 

Tags have properties, parameters have properties. This concept will be expanded upon in JavaScript and Python, when we learn that properties can have properties, which have properties, that can also have properties which ... you get the picture ...

And these things can be, and are, *nested* within each other --- which relates to the importance of *closing* our parameter, functions and other mechanics, properly.

## JavaScript

JavaScript is a tricky beast because there is no one correct way to write it!

Ok, granted, it is starting to formalize, and has been doing so for a while. Regardless there remain a variety of custom structures used across the many varied libraries.

When we see it in our code, it will often be in the form of specials VueJS structures. Though there are many structures and concepts which are consistent across libraries.

```html
<script src="bootstrap.min.js"></script>
<script src="jquery.min.js"></script>

<script>
$('.jobLabel a').on('click', function(){

    var target = $(this).attr('rel');
    
    $("#"+target).show().siblings("div").hide();   
});
</script>

```

There are a few unusual things to notice here, the first of which is the use of the dollar-sign ~

We are only able to use this because of the libraries we pulled in earlier, *jquery.min.js* and *bootstrap.min.js*. 

Really only the jQuery script, but moving on ...

This is a complex-looking function, so let's disassemble it and see what's going on:

#### The $

The dollar-sign allows us to access the DOM, document object manager. So inside the DOM, we can look for tags!

Parenthesis contain things, and contained in these first parethesis' are list of items separated by a space, *.jobLabel* and *a*. 

We recognize the syntax of *.jobLabel* from CSS, and determine that it indicates a class, while the *a* is an html tag!

> 1) We have an object on the webpage/DOM with which we are working!

> `$('.jobLabel a')`
    
#### Reacting

Now that we have the object discerned, what is happening with it?!

There is a period after our parenthetical statement, followed by the word *on*, and another set of parenthesis.

> 2) We have an action, or set of actions.

> `.on('click',function`

In JS5 this all means, when the our DOM object is clicked, *on-click*, the following function should happen.
    
#### Acting

Within this function we declare and define a variable calling it *target*, and set this variable using our dollar-sign, while specifying a specific propery attribute to read, *rel*.

We then use this property attribute value to identify a third object. 

We *show* this item, while *hiding* all of its *siblings*.

> 3) Within the function we are able to affect,and change, the DOM.

> `var target = $(this).attr('rel');`
    
> `$("#"+target).show().siblings("div").hide();`

Notice how each line ends with a semi-colon.

And how each parenthetical statement, or bracketed statement, is closed.

### Variables

Variables are a way of storing information, and using it elsewhere. 

And so there are various types of variables, *strings*, *lists*, and *dictionaries* to start. 


#### Strings

Strings are a collection of characters enclosed by quotations, single or double, it matters not which.

```javascript
var target = 'some string';
```

#### Boolean

Boolean values are *true/false* values; one or the other.

This is translated into machine code as either a `0` or `1`.

```javascript
var target = true;
# or
var target = 1;
```

#### Lists

Lists are a collection of objects separated by commas and enclosed by brackets `[]`. 

Lists start at 0.

The objects do not all need to be of the same type.

```javascript
var target = ['some', 'list', 'of', 'strings', ['with another list', 'within',], true,];
```

#### Dictionaries

Dicionatires are a collection of *key-value pairs*.

Let's break that down, the *key* is defined by a string, and separated from the *value* by a colon `:`. 

Each *key/value pair* is then separated by comma `,`.

This is all enclosed within *curly brackets* `{}`.

```javascript
var target = {'text': 'some string', 'published': false, 'colors': ['red', 'blue',],};
```

#### If-statements

```javascript

if (x < y) {
  var display = "Welcome to SF Coding 101";
} else if (x < z) {
  var display = "Welcome to SF Coding 102";
} else {
  var display = "We haven't planned this far ahead yet :/ Check with class mates to see how they are, porfa";
}
```

Read it out-loud:

If *x* is *less than y*, the variable *display* equals a string.

*Else, if x* is not less than *y* and is *less than z*, *display* equals a different string.

If *x* is larger than them all, using the *else* statement, *display* is set to equal a third string.

### Inputs and outputs

Super fun, but at day's end all we have is an *Element* of some type *html* ...

We need to learn how to dig deeper into our webpages!

When we used the `html.fromstring` tool, something else happened; our new *base* object has its own properties and functions.

The tool `xpath` is what allows us to traverse the DOM in a scraper!

Try running the following code.

````python
import requests
from lxml import html

site = requests.get("https://cultureclap.github.io")
base = html.fromstring(site.text)
print(base.xpath('.//*/h1/a/text()'))
````

**So what is happening here?!**

`page.xpath('.//*/h1/a/text()')`

We recognize the `h1` and `a` tags, they are html tags, in theory.

`xpath` is new, and so is the `text()` part.

`xpath('')` is the structure from which we build.

In our particular use `xpath('.//')` will be how we start string the function `xpath` consumes.

> What this assures is that we are always starting from the base of our object.

So let's read this command out-loud:

From the `page` object we are use `xpath` to traverse the DOM, starting from the base of the `page`.

The `*` denotes a wildcard.

Looking through *every* object, if there is an `h1` nested underneath, with an `a` tag, we want to do something called `text()`.

> When words are followed by parenthesis `()`, it often signifies a function.

In this case, when `text()` is called, the code is asking to read any available text that may be nexted in the previous tag.
